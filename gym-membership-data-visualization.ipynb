```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/gym-membership-dataset/gym_membership.csv')
```

```python
data.tail()
```

```python
data.dtypes
```

# Bar chart of gym membership types

```python
data['abonoment_type'].value_counts().plot(kind='bar', color='blue')

plt.title("Gym Membership Types")
plt.xlabel("Membership Type")
plt.ylabel("Count")
plt.show()
```

# Histogram of time in gym

```python
plt.hist(data['avg_time_in_gym'], bins=15, color='blue', edgecolor='black')
plt.title("Distribution of Average Time in Gym")
plt.xlabel("Time in Gym (minutes)")
plt.ylabel("Frequency")
plt.show()
```

# Scatter plot: Age vs visits per week

```python
plt.scatter(data['Age'], data['visit_per_week'])
plt.grid(True)
plt.title("Age vs Visits per Week")
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.show()
```

# Box plot: Gender vs Age

```python
sns.boxplot(data=data, x='gender', y='Age', palette='Set2')
plt.title("Age Distribution by Gender")
plt.show()
```

# Count plot: Days per week

```python
sns.countplot(data=data, x='days_per_week', palette='viridis')
plt.title("Count of Days per Week in Gym")
plt.show()
```

# Bar plot: Average visits by membership type

```python
sns.barplot(data=data, x='abonoment_type', y='visit_per_week', estimator='mean', palette='coolwarm')
plt.title("Average Visits per Week by Membership Type")
plt.ylabel("Average Visits per Week")
plt.show()
```
