```python
import numpy as np
import pandas as pd

data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
data.head(10)
```

```python
cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
for col in cols:
    print(f"{col}: {data[col].unique()}")
```

```python
data.describe()
```

```python
math = data['math score'].values
reading = data['reading score'].values
writing = data['writing score'].values
```

```python
print("Math\nMean:", np.mean(math))
print("Median:", np.median(math)) 
print("Std:", np.std(math))
print("Min:", np.min(math))
print("Max:", np.max(math))
```

```python
print("Reading\nMean:", np.mean(reading))
print("Median:", np.median(reading)) 
print("Std:", np.std(reading))
print("Min:", np.min(reading))
print("Max:", np.max(reading))
```

```python
print("Writing\nMean:", np.mean(writing))
print("Median:", np.median(writing)) 
print("Std:", np.std(writing))
print("Min:", np.min(writing))
print("Max:", np.max(writing))
```

```python
math_above_80 = np.sum(math > 80)
print("\nStudents with math > 80:", math_above_80)
```

```python
below_50_any = np.sum((math < 50) | (reading < 50) | (writing < 50))
print("Students with any score < 50:", below_50_any)
```

```python
overall_avg = np.mean([math, reading, writing])
print("Overall average score across all subjects:", overall_avg)
```

```python
gender_group = data.groupby('gender').agg(
    avg_math=('math score', 'mean'),
    avg_reading=('reading score', 'mean'),
    avg_writing=('writing score', 'mean'),
    student_count=('gender', 'count')
)
print("1. Group by Gender:\n", gender_group, "\n")
```

```python
testprep_group = data.groupby('test preparation course').agg(
    avg_math=('math score', 'mean'),
    avg_reading=('reading score', 'mean'),
    avg_writing=('writing score', 'mean')
)
print("2. Group by Test Preparation Course:\n", testprep_group, "\n")
```

```python
parent_group = data.groupby('parental level of education').agg(
    avg_math=('math score', 'mean')
).sort_values(by='avg_math', ascending=False)
print("3. Group by Parental Education (Ranked by Math):\n", parent_group, "\n")
```

```python
lunch_group = data.groupby('lunch').agg(
    avg_math=('math score', 'mean'),
    avg_reading=('reading score', 'mean'),
    avg_writing=('writing score', 'mean'),
    student_count=('lunch', 'count')
)
print("4. Group by Lunch Type:\n", lunch_group, "\n")
```

```python
avg_math_by_gender = data.groupby('gender')['math score'].mean()
math_diff = avg_math_by_gender['male'] - avg_math_by_gender['female']
print("1. Average Math Score by Gender:\n", avg_math_by_gender)
print("   Difference (Male - Female):", math_diff, "\n")
```

```python
overall_avg = {
    'Math': data['math score'].mean(),
    'Reading': data['reading score'].mean(),
    'Writing': data['writing score'].mean()
}
overall_avg_sorted = dict(sorted(overall_avg.items(), key=lambda x: x[1], reverse=True))
print("2. Overall Subject Averages (Highest to Lowest):\n", overall_avg_sorted, "\n")
```

```python
data['total score'] = data['math score'] + data['reading score'] + data['writing score']
parent_impact = data.groupby('parental level of education')['total score'].mean().sort_values(ascending=False)
print("3. Average Total Score by Parental Education (Ranked):\n", parent_impact, "\n")

print("   Highest performing parental education level:", parent_impact.idxmax())
```
